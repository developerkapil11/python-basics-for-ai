```python
class Employee:
    print('Hello World!')
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary
    


emp1 = Employee("kapil", 38, 50000)
emp2 = Employee("Maahi", 40, 70000)

print(emp1.name, emp2.name, emp1.age, emp2.age, emp1.salary, emp2.salary)
```

```python
class Car:

    discount = 10

    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price

    def show_details(self):
        print(self.brand, self.model, self.price)

    def show_price_after_discount(self):
        discount_price = (self.price * 10) /100
        print(discount_price)
        self.price = self.price - discount_price
        print(self.price)
        


car1 = Car('BMW', 'GT200', 12000000)
car2 = Car('TATA', 'Nexon', 1500000)
car3 = Car('KIA', 'Seltos', 1200000)

car1.show_details()
car2.show_details()
car3.show_details()

car1.show_price_after_discount()
```

```python
class Dog:
    def __init__(self, breed, name):
        self.breed = breed
        self.name = name
    
    def show_dog(self):
        print(f"{self.name}'s breed is = {self.breed}")

class Cat:
    def __init__(self, name, color):
        self.name = name
        self.color = color
    
    def show_cat(self):
        print(f"{self.name} cat has {self.color} color")

dog1 = Dog("Labradoar", "Jerry")
dog2 = Dog("Desi", "Kaalu")

cat1 = Cat("marry", "White")
cat2 = Cat("tom", "black")

dog1.show_dog()
dog2.show_dog()
cat1.show_cat()
cat2.show_cat()
```
