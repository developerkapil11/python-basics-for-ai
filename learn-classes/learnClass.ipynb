```python
class Car:

    discount = 10

    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price

    def show_details(self):
        print(self.brand, self.model, self.price)

    def show_price_after_discount(self):
        discount_price = (self.price * 10) /100
        print(discount_price)
        self.price = self.price - discount_price
        print(self.price)
        


car1 = Car('BMW', 'GT200', 12000000)
car2 = Car('TATA', 'Nexon', 1500000)
car3 = Car('KIA', 'Seltos', 1200000)

car1.show_details()
car2.show_details()
car3.show_details()

car1.show_price_after_discount()
```

```python
class Dog:
    def __init__(self, breed, name):
        self.breed = breed
        self.name = name
    
    def show_dog(self):
        print(f"{self.name}'s breed is = {self.breed}")

class Cat:
    def __init__(self, name, color):
        self.name = name
        self.color = color
    
    def show_cat(self):
        print(f"{self.name} cat has {self.color} color")

dog1 = Dog("Labradoar", "Jerry")
dog2 = Dog("Desi", "Kaalu")

cat1 = Cat("marry", "White")
cat2 = Cat("tom", "black")

dog1.show_dog()
dog2.show_dog()
cat1.show_cat()
cat2.show_cat()
```

```python
class BankAccount:

    bankName = "HDFC Bank"

    def __init__(self, accountHolder, balance):
        self.accountHolder = accountHolder
        self.balance = balance
    
    @classmethod
    def change_bank_name(cls, bank_name):
        cls.bankName = bank_name


person1 = BankAccount("Kapil", 2000000)
person2 = BankAccount("Maahi", 2500000)

BankAccount.change_bank_name("SBI BANK")

person2.bankName = "ICICI BANK"

print(person1.accountHolder, person1.balance, person1.bankName)
print(person2.accountHolder, person2.balance, person2.bankName)
    
```

```python
class Temprature:
    unit = 'celsius'

    @classmethod
    def change_unit(cls,new_unit):
        cls.unit = new_unit
    
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        F = (celsius * 9/5) + 32
        return F

print(Temprature.celsius_to_fahrenheit(25))
```

```python
class Citizen:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_csv(cls, data):
        name, age = data.split(",")
        return cls(name, int(age))
    
person1 = Citizen.from_csv("kapil,35")
person2 = Citizen.from_csv("maahi,30")

print(person1.name, person1.age)
print(person2.name, person2.age)
```

```python
class Employee:
    def __init__(self, name, age, salary, email):
        self.name = name
        self.age = age
        self.salary = salary
        self.email = email
    
    def show_employee_detail(self):
        print(self.name, self.age, self.salary, self.email)

class Developer(Employee):
    def __init__(self, name, age, salary, email,  language):
        super().__init__(name, age, salary, email)
        self.language = language

    def show_dev_details(self):
        print(f"this is dev class {self.name} and know the language = {self.language}")

class Manager(Employee):
    def __init__(self, name, age, salary, email, total_team_count):
        super().__init__(name, age, salary, email)
        self.total_team_count = total_team_count

    def show_employee_detail(self):
        print(self.name, self.age, self.salary, self.email, self.total_team_count)

class Tester(Employee):
    pass

dev1 = Developer("kapil", 35, 100000, "kplhans@gmail.com", "Python")
dev1.show_employee_detail()

mgr1 = Manager('Aashish', 32, 200000, 'aashish@gmail.com', 15)
mgr1.show_employee_detail()

dev1.show_dev_details()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def show(self):
        print(self.name, self.age)

per1 = Person("maahi", 30)
per1.show()

class ClgStudent(Person):
    def __init__(self, name, age, course):
        super().__init__(name, age)
        self.course = course

    def show(self):
        super().show()
        print(self.course)

clgStu1 = ClgStudent("kavish", 15, "AI/ML Engineer")
clgStu1.show()

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject
    
    def show(self):
        super().show()
        print(self.subject)

tchr1 = Teacher("Meenakshi", 45, "Software Engineering")
tchr1.show()
```
