```python
# AbstractMethod

# ABC guarantees that every subclass provides these methods before it can be instantiated.
from abc import ABC, abstractmethod 

class Payment(ABC):

    @abstractmethod
    def pay(self, amount):
        pass

class Stripe(Payment):
    def pay(self, amount):
        print('Stripe-', amount)

class Paypal(Payment):
    pass


#pay1 = Paypal()   # Can't create object becuase AbstractMethod in parent and it should have pay method inside it Error:- Can't instantiate abstract class Paypal without an implementation for abstract method 'pay'


#pay1.pay(1000)

strp1 = Stripe()
strp1.pay(5000)
```

```python
from abc import ABC, abstractmethod
class Duck(ABC):

    @abstractmethod
    def sound(self):
        print("Quack!")

class Human(Duck):
    def sound(self):
        print("Helloooo")

class Speaker(Duck):
    pass

def make_sound(obj):
    obj.sound()

make_sound(Human())
# make_sound(Speaker()) # Can't instantiate abstract class Speaker without an implementation for abstract method 'sound'
```

```python
# Another example of abstract class (with 2 abstract methods)

from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        print("Start!!!!!!!")

    @abstractmethod
    def stop(self):
        print('STOP........')

class Car(Vehicle):
    def start(self):
        print('Car Start Car !!!')
    
    def stop(self):
        print('Car Stop Car ...')

class Bike(Vehicle):
    def stop(self):
        print('Bike Stop Bike ...')

car1 = Car()
car1.start()
car1.stop()

# bike1 = Bike() #  Can't instantiate abstract class Bike without an implementation for abstract method 'start'
# We need to declare both the functions in class Bike (because both functions are declared in parent class)
# bike1.start()
```

```python
# Protocol in python

from typing import Protocol

class Logger(Protocol):

    def log(self, message: str):
        print("main Logger: ", message)

class ConsoleLogger():

    def logss(self, message):
        print("console: ", message)

class DatabaseLogger:

    def log(self, message):
        print("Saving: ", message)

def make_logger(logger: Logger):
    logger.log("Helloooo Bro!!!!")

   

make_logger(DatabaseLogger())
# make_logger(ConsoleLogger())
con1 = ConsoleLogger()
con1.logss("Hello Mam")
```

```python
class SecondLargest:
    def secondLargest(self, nums: list[int]):
        if not isinstance(nums, list):
            return []
        
        firstLar = float('-inf')
        secondLar = float('-inf')

        for i in range(len(nums)):
            if nums[i] > firstLar:
                secondLar = firstLar
                firstLar = nums[i]
            elif firstLar > nums[i] > secondLar:
                secondLar = nums[i]
        
        return secondLar
    
obj = SecondLargest()
obj.secondLargest([2,5,66,7,8,99,11,234, 101])

                
```
